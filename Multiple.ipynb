<h1>Multiples Lines, Single Plot</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
```

<h2>random low and high temperature data</h2>

```python
data_set_size = 15
low_mu, low_sigma = 50, 4.3
low_data_set = low_mu + low_sigma * np.random.randn(data_set_size)
high_mu, high_sigma = 57, 5.2
high_data_set = high_mu + high_sigma * np.random.randn(data_set_size)

days = list(range(1, data_set_size + 1))
```

```python
plt.plot(days,low_data_set)
```

```python
plt.plot(days, low_data_set, days, high_data_set)
plt.show()
```

```python
plt.plot(days,low_data_set,days,low_data_set,"vm",days,high_data_set,days,high_data_set,"^k")
plt.show()
```

```python
plt.plot(days,low_data_set,days,high_data_set)
plt.xlabel('Day')
plt.ylabel('Temperature: degress Farenheit')
plt.title('Randomized temperature data')
plt.show()
```

```python
t1 = np.arange(0.0,2.0,0.1)
t2 = np.arange(0.0,2.0,0.01)

#note that plot returns a list of lines. The "L1" = plot usage extrct the
#first element of the list into L1 using tuple unpacking. So L1 is  a Line 2D
#instance, not a sequence of lines.
L1 = plt.plot(t2,np.exp(-t2))
L2,L3 = plt.plot(t2,np.sin(2*np.pi*t2), '--go',t1,np.log(1+t1),'.')
L4 = plt.plot(t2, np.exp(t2)*np.sin(2*np.pi*t2),'rs-.')

plt.legend((L2, L4),('oscilltory','damped'),loc='upper right',shadow = True)#error showing "damped" in legend
plt.xlabel('time')
plt.ylabel('volts')
plt.title('Damped oscillation')
plt.show()
```
